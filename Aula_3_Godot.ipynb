```python
from re import A
# 8. Sistema de login com tentativas
# Crie um pseudocódigo que:
# Permite até 3 tentativas de login antes de bloquear o usuário.

usuario_correto = "admin"
senha_correta = "3423428423950938"

tentativas = 0
max_tentativas = 3

while tentativas < max_tentativas:
  usuario = input("login: ")
  senha = input("senha: ")

  if usuario == usuario_correto and senha == senha_correta:
    print("acesso permitido")
    break
  else:
    tentativas += 1
    print(f"algo esta errado. voce tem {max_tentativas-tentativas} tentativas")

if tentativas == max_tentativas:
  print("bloqueado")
```
